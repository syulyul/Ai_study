```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 10))
for i in range(100):
    plt.subplot(10, 10, i+1)
    plt.subplots_adjust(
        wspace=0.5, hspace=0.5
    )
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap = plt.cm.binary)
    plt.xlabel(class_names[y_train[i][0]])
#plt.imshow(x_train[0], 'gray')
plt.show()

```
